```python
def probable_kmer(dna,k,profile):
    most_probable_kmer=""
    dic2={}
    val= float('-inf')
    dic={'A':0,'C':1,'G':2,'T':3}
    for i in range (len(dna)-k+1):
        kmer=dna[i:i+k]
        current_val=0
        for j in range(k):
            current_val+=profile[dic[kmer[j]]][j]
        dic2[kmer]=current_val    
        if current_val>val:
            val=current_val
            most_probable_kmer=kmer
    return most_probable_kmer,dic2       
    
```

```python
dna = "CTTAATGGTTAGGAGTGACTTGATCTGAACCGATCCACGGCACTAGTTCCCGCCCATCACGCTCAACGACCGTGTGTCATCGTGATCAAGGTTTTCGTTGCTATTGGTGCTCGCAGATATGCGCATGCCAATGTGTGCTAGTGCTAGGAAGATTAACTAGAATGTGCTAGACCAGCTCGGCGTTGTATTCCGGACGTAGG"
k= 7
profile = [
    [0.214, 0.393, 0.0, 0.25, 0.179, 0.214, 0.321],
    [0.286, 0.143, 0.393, 0.25, 0.357, 0.286, 0.25],
    [0.286, 0.357, 0.321, 0.179, 0.214, 0.214, 0.357],
    [0.214, 0.107, 0.286, 0.321, 0.25, 0.286, 0.071]
]
ans,dic2=probable_kmer(dna,k,profile)

for key,val in dic2.items():
    print(f"{key}: {val}")
print("probable:"+ans)    
```
